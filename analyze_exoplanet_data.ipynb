```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("exoplanet_data.csv",comment = "#")
pd.read_csv("exoplanet_data.csv",comment = "#")
```

```python
print(data.shape) # 567 rows, 92 columns
print(data.columns) # pl_orbperlim, pl_bmasseerr2, decstr, st_tefflim, sy_gaiamag
# Planet name --> pl_name†, planet name used most commonly.
# Host name --> hostname†, stellar name used commonly.
# Discovery method --> discoverymethod†, The method used to first identiy the planet
# Spectral type --> st_spectype, Classifying stars based on their spectral characteristics.
# Equalibrium temerpature --> pl_eqt†, the effective temperature for a planet that absorbs heat from its system or to match its luminoscity if it where able to absorb energy ideally (black body).
# Planet Mass --> pl_massj (jupiter), amount of matter contained in the planet measured in jupiters 

data_col = pd.DataFrame(data[["pl_name","hostname","discoverymethod","st_spectype","pl_eqt","pl_bmassj"]])
data_col = data_col.dropna()
print(data_col.shape) # 67 rows, 6 columns
print(f"Head: {data_col.head()}")
print(f"Tail: {data_col.tail()}")
print(data_col.columns)
```

```python
print(data_col["discoverymethod"].value_counts())
# Discovery methods include: Transit (52), Radial Velocity (13) and Imaging (2).
# Transit is repersented the most: Observing when a light dip occurs due to path taken where the exoplanet passes between us and a star
# Radial velocity: Observign the changes in a star's motion in orbit of a planet effecting the light it gives off
# Imaging if the least repersented in the data: oberving imaging of the exoplanet.
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(data = data_col, x= 'pl_bmassj', y = 'pl_eqt', hue = 'discoverymethod')
plt.title("Planet Mass versus Equalibrium temperature")
plt.xlabel("Planet Mass (Jupiters)")
plt.ylabel("Equalibrium temperature (Kelvin)")
plt.show()
```

```python
plt.figure(figsize = (8,6))
x = np.log(data_col["pl_bmassj"])
y = np.log(data_col["pl_eqt"])
df_removed = data_col.drop(['pl_bmassj', 'pl_eqt'], axis=1)
df_removed["pl_bmassj"] = x
df_removed["pl_eqt"] = y
sns.scatterplot(data = df_removed, x= 'pl_bmassj', y = 'pl_eqt', hue = 'discoverymethod')
plt.title("Logarithmic plot: Planet Mass versus Equalibrium temperature")
plt.xlabel("Planet Mass (log(Jupiters))")
plt.ylabel("Equalibrium temperature (log(Kelvin))")
plt.show()
# The planets found by imaging are found to have relatively more mass than other observed planets.
```

```python
mask = data_col["pl_bmassj"] > 5
over_five = data_col[mask]
print(over_five["pl_eqt"].min())
coldest = over_five["pl_eqt"] == 275.0
coldest_planet = over_five[coldest]
coldest_planet = coldest_planet.dropna()
print(coldest_planet)

# name = eps Ind A b, index = 566, temp = 275.0, mass = 6.31 jupiters, orbits a similar star to our sun (K5 V) a main sequency star.
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(data = df_removed, x= 'pl_bmassj', y = 'pl_eqt', hue = 'discoverymethod')
plt.scatter( np.log(coldest_planet["pl_bmassj"]),np.log(coldest_planet["pl_eqt"]), c = "red", s = 40, label = "K5 V" )
plt.title("Logarithmic plot: Planet Mass versus Equalibrium temperature")
plt.xlabel("Planet Mass (log(Jupiters))")
plt.ylabel("Equalibrium temperature (log(Kelvin))")
plt.legend()
plt.show()
```
