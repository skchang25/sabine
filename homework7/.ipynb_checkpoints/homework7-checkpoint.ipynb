```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
amplitude = np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])
def make_sin_wave(x,A,w):
    return A*np.sin(w*x)
fig,axes = plt.subplots(1, 3, figsize=(25,5))

for w in frequency:
    y = make_sin_wave(x, A=1, w=w)  # fix amplitude at 1
    axes[1].plot(x, y, label="frequency varied")
axes[1].set_title("Frequency varied")

for A in amplitude:
    y = make_sin_wave(x,A,w=1)
    axes[0].plot(x,y,label = "amplitude varied")
axes[0].set_title("Amplitude varied")

for w,A in zip(frequency, amplitude):
    y = make_sin_wave(x,A,w)
    axes[2].plot(x,y)
axes[2].set_title("Frequency and Amplitude varied")
    
plt.show()
```
