```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 5)
amplitude = np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])
def make_sin_wave(x,A,w):
    return A*np.sin(w*x)
fig,axes = plt.subplots(1, 3, figsize=(25,5))
axes[0].plot(x,make_sin_wave(x,amplitude,1))
axes[0].set_title('amplitude varied over time')
axes[1].plot(x,make_sin_wave(x,1,frequency))
axes[1].set_title("frequency varied over time")
axes[2].plot(x,make_sin_wave(x,amplitude,frequency))
axes[2].set_title("sin wave: frequency and amplitude varied")
plt.show()
```
