```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sin_wave(x,A,w):
    y = np.array([])
    y = A*np.sin(w*x)
    return y
```

```python
x = np.linspace(0,2*np.pi, num = 1000)
amplitude = np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])
fig, axs = plt.subplots(25,5)
axs[0].plot(x,amplitude)
axs[0].set_title('amplitude')
axs[1].plot(x,frequency)
axs[1].set_title("frequency")
axs[2].plot(x,make_sin_wave(x,amplitude, 2*np.pi*frequency))
axs[2].set_title("sin wave")
plt.show()
```
