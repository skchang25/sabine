```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt",
skiprows=1)
```

```python
print(data.shape) # 2642 rows
```

```python
magnitude = data[:,:1]
temperature = data[:,1:2]
print("minimum magnitude (brightest star): ", magnitude.min()) # aroung the magnitude of the sun
print("minimum temperature: ", temperature.min()) # around AB Doradus C
print("maximum magnitude (dimmest star): ", magnitude.max()) # just below sequence and just above the sun
print("maximum temperature: ", temperature.max()) # just around Main
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
fig,axes = plt.subplots(1, 1, figsize = (8,6))
plt.scatter(log_temperature,log_magnitude)
#plt.plot(log_magnitude.min(), color = "black")
#plt.plot(np.log(magnitude.max()), color = "purple")
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
```
